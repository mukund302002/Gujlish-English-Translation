```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("mukundagarwalla2002/nllb-guj_eng-translation")
model = AutoModelForSeq2SeqLM.from_pretrained("mukundagarwalla2002/nllb-guj_eng-translation")
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

def translate(text, src_lang="guj_Gujr", tgt_lang="eng_Latn", max_length=256):
    tokenizer.src_lang = src_lang

    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=max_length
    ).to(device)

    with torch.no_grad():
        output_tokens = model.generate(
            **inputs,
            forced_bos_token_id=tokenizer.convert_tokens_to_ids(tgt_lang),
            max_length=max_length,
            num_beams=5
        )

    return tokenizer.decode(output_tokens[0], skip_special_tokens=True)
```

```python
text = "Hu college jato hato ane exam bahu tough hato."
print(translate(text))
```
